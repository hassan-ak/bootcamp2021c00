#### P32 Conditional Statements
* if
* elif
* else

```python
x = 10
y = 12
if(x<y):
    print("x is less than 5")
    print("on line 6")
elif(x>y):
    print("x is grater than y")
else:
    print('x and y are equal')
```

```python
marks = 70
if (marks<40):
    print("Fail")
elif(40<=marks<=60):
    print("Average")
else:
    print("Well Done")
```

#### P34 While Loop

```python
count = 0
while(count<5):
    print(count)
    count = count+1
print("Good Bye")
```

```python
rank=5
while(rank !=12):
    print("Rank is ",rank)
    rank+=1
```

```python
input_number = 0
correct_answer = 13
while (input_number != correct_answer):
    input_number = int(input("Input Number: "))
print("Exit: Congratulations. You Made it!")
```

#### P035 For Loop
* it iterates over a range

```python
fruits = ["Banana", "Apple", "Grapes"]
for fruit in fruits:
    print(fruit)
```

```python
list1 = [1,2,3,"Python"]
for i in list1:
    print(i)
```

#### P036 Nested Loops

```python
count = 1
for i in range(10):
    print(str(i)*i)
    for j in range(0,i):
        count = count+1
print(count)
```

#### P037 Control Statements
* break statement
* continue statement
* pass statement

```python
for i in range(1,11):
    if i == 5:
        continue
    print(i)
```

```python
for i in range(1,11):
    if i == 5:
        break
    print(i)
```

```python
a = 1
entered_card = "no"
retry = 0
pin = "1234"
current_balance = "10000"
while(a == 1):
    if entered_card != "e":
        print("*************************************")
        entered_card = input("Enter your card (e) :")
        print("*************************************")
    if entered_card == "e":
        entered_pin = input("Enter your pin")
        if entered_pin == pin:
            while (entered_pin == pin): 
                print("*************************************")
                print("** Select from the following menu  **")
                print("** 1 ------> Withdraw Amount       **")
                print("** 2 ------> Balance Inquiry       **")
                print("** 3 ------> Pay Amount            **")
                print("** 4 ------> Take your Card       **")
                print("*****************************************")
                selection = input("select from Above Menu :")
                if(selection == "1"):
                    selected_amount = input("Kindly enter amount to withdraw :")
                    print("Amount withdrawn : ", selected_amount)
                if(selection == "2"):
                    print("Your current balance is : ", current_balance)
                if(selection == "3"):
                    selected_amount = input("Kindly enter amount to pay :")
                    print("Payed amount : ", selected_amount)
                if(selection == "4"):
                    print("Hope to see you soon")
                    retry = 0
                    entered_pin = "1"
                    entered_card = "a"
        else:
            retry = retry + 1
            print("Kindly enter correct pin (Wrong pin = " ,retry,")")
            if(retry == 3):
                print("*****************************************")
                print("** You Entered wrong pin for 3 times.  **")
                print("** Card Returned.                      **")
                print("*****************************************")
                retry = 0
                entered_card = "no"
    if (entered_card == "123456789"):
        break
```

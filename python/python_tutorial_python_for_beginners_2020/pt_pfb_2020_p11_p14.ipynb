## P12 If Statements

```python
temperature = 35
if temperature>30:
    print("It's a hot day")
    print("Drink Plenty of water")
```

```python
temperature = 25
if temperature>30:
    print("It's a hot day")
    print("Drink Plenty of water")
```

```python
temperature = 35
if temperature>30:
    print("It's a hot day")
    print("Drink Plenty of water")
print("Done")
```

```python
temperature = 0
if temperature>30:
    print("It's a hot day")
    print("Drink Plenty of water")
elif temperature >20: # (20,30)
    print("It's a nice day")
elif temperature >10: # (20,30)
    print("It's a bit cold")
else:
    print("It's Cold")
print("Done")
```

## P 13 Exercise

```python
weight = float(input("Weight : "))
unit = input("(K)g or (L)bs : ").lower()
if unit == "l":
    print("Weight in Kg : " + str(weight*0.45))
elif unit == "k":
    print("Weight in Lbs : " + str(weight/0.45))
else:
    print("Kindly select correct units.")
```
